```python
# DIESEN CODEBLOCK NICHT ÄNDERN ------------------------------------------------

def start_quiz(sheet_name="Sample", num_questions=None, points_per_answer=1):
  import pandas as pd

  sharing_url = "https://docs.google.com/spreadsheets/d/17Bz-vMzxrNk6-WDuE3iXhYbfFEHkAwbjwAGaEEWNG7I/edit?usp=sharing"

  # You can find the sheet id in the url of your file behind “d/”, copy it from your browser and paste it into your code.
  # The sheet name is the name you gave your sheet.
  spreadsheet_id = sharing_url.split('/')[-2]
  url = f"https://docs.google.com/spreadsheets/d/{spreadsheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"
  data = pd.read_csv(url)

  def ask_questions(data=data, num_questions=None, points_per_answer=1):
    score = 0

    if num_questions:
      n = min(num_questions, len(data))
      data = data.sample(n=n).reset_index(drop=True)
    else:
      data = data.sample(n=len(data)).reset_index(drop=True)

    for i, row in data.iterrows():
      print(f"Frage {i+1}/{len(data)}")
      frage = row["Frage"]
      antworten = row.filter(regex="Antwort \d+").to_dict()

      # {int: str}
      antworten = {
          int(k.replace("Antwort ", "")): str(v).replace('"', '')
          for k, v in antworten.items()}

      # str
      gültige_antworten = ', '.join([str(k) for k in antworten.keys()])

      # str
      richtige_antwort = row.filter(regex="ichtig").iloc[0]

      prompt = f"{frage}:\n" + ''.join([f"  {k}) {v}\n" for k, v in antworten.items()])
      antwort = input(prompt)

      # check for int in keys
      while (not antwort.isnumeric()) or int(antwort) not in antworten.keys():
        print(f"Ungültige Antwort. Gültige Antworten sind folgdende: {gültige_antworten}")
        antwort = input()

      antwort = int(antwort)
      if antwort == richtige_antwort:
        score += points_per_answer
        print("Richtig!")
      else:
        print(f"Falsch. Die richtige Antwort ist:\n  {richtige_antwort}) {antworten[richtige_antwort]}")

      print("---")

    print(f"Ergebnis: {score}/{len(data)*points_per_answer}")
    return None

  ask_questions(num_questions=num_questions, points_per_answer=points_per_answer)
  return None
# ------------------------------------------------------------------------------
```
